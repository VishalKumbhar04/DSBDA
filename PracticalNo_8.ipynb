```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset('titanic')
dataset.head()
```

```python
pt = [20,25,28,30,80,40,45,50,60,25]
pt1 = [30,15,80,48,50,60,70,39,80,99]

plt.plot(pt)
plt.plot(pt1)
plt.show()
```

```python
plt.plot(pt, c="r", ls="-.", marker="X")
plt.plot(pt1, c="y", ls="-", marker="p")

plt.xlabel("Sample Points")
plt.ylabel("Values")
plt.title("Values vs Points")
plt.legend(["A", "B"])

plt.show()
```

```python
sns.histplot(dataset['age'], bins=10, kde=True)
plt.show()
```

```python
sns.jointplot(x=dataset['age'], y=dataset['fare'], kind='scatter')
plt.show()
```

```python
sns.jointplot(x=dataset['age'], y=dataset['fare'], kind='hex')
plt.show()
```

```python
sns.rugplot(dataset['fare'])
plt.show()
```

```python
sns.barplot(x='sex', y='age', data=dataset)
plt.show()
```

```python
sns.barplot(x='sex', y='age', data=dataset, estimator=np.std)
plt.show()
```

```python
sns.countplot(x='sex', data=dataset)
plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=dataset)
plt.show()
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue='survived')
plt.show()
```

```python
sns.violinplot(x='sex', y='age', data=dataset)
plt.show()
```

```python
sns.stripplot(x='sex', y='age', data=dataset, jitter=True)
plt.show()
```

```python
sns.swarmplot(x='sex', y='age', data=dataset)
plt.show()
```

```python
corr = dataset.corr(numeric_only=True)
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.show()
```
