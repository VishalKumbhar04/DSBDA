```python
rmse = np.sqrt(mse)
print("RMSE:", rmse)
```

```python
y_mean = np.mean(y_test)
y_mean
```

```python
SST = np.sum((y_test - y_mean)**2)
print("SST:", SST)
```

```python
SSR = SST - SSE
print("SSR:", SSR)
```
