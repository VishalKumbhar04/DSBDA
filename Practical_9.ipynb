```python
sns.barplot(x='sex', y='age', hue='sex', data=dataset)
```

```python
sns.barplot(x='sex', y='age',hue = 'sex', data=dataset, estimator=np.std)
```

```python
sns.countplot(x='sex',hue = 'sex', data=dataset)
```

```python
sns.boxplot(x='sex', y='age',hue = 'sex',data=dataset)
```

```python
sns.boxplot(x='sex', y='age', data=dataset, hue="survived")
```

X-axis: Sex (male vs female)
Y-axis: Age distribution
Hue: Survival (0 = not survived, 1 = survived)
Survival difference by gender
Females (survived = 1) → Higher count and wider box
Males (survived = 0) → Larger non-survivor group
