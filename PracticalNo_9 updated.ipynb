```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset = sns.load_dataset("titanic")
dataset.head()
```

```python
dataset.describe()
```

```python
dataset.info()
```

```python
dataset['age'] = dataset['age'].fillna(dataset['age'].mean())
```

```python
dataset['age'].isnull().sum()
```

```python
dataset.info()
```

```python
sns.boxplot(x='sex', y='age', hue='survived', data=dataset)

plt.title('Age Distribution of Survived and Unsurvived Data')
plt.xlabel('Sex')
plt.ylabel('Age')

plt.show()
```

```python
sns.countplot(x='sex', hue='survived', data=dataset)
plt.title('Survival Count based on Gender')
plt.show()
```

Observation:- The middle age (median) of both survived and not survived people is
around 28 to 30 years for both males and females. For males, the age of survived
and not survived people is almost the same, mostly between 20 to 35 years. This
means age does not affect male survival much, because both groups look similar.
For females, most of the survived people are also between 20 to 35 years. Females
who did not survive are slightly younger compared to those who survived. The age
range (spread) is a bit wider in females, so their ages vary more. There are some
outliers (extreme values): Very small children (below 5 years) Very old people
(above 60–70 years) Females have a slightly better chance of survival compared to
males.
